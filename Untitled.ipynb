```python
import requests
import pandas as pd
from pandas import json_normalize

def get_json_response(url):
    response = requests.get(url)
    response.raise_for_status()
    return response.json()

def normalize_json_response(json_data):
    # Normalize the JSON data into a flat DataFrame
    df = json_normalize(json_data['results'])

    return df

url = 'https://api.mercadolibre.com/sites/MLA/search?category=MLA1000&offset=50'
json_data = get_json_response(url)
df = normalize_json_response(json_data)

df
```
